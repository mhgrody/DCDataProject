```python
#Shows the statistics of all the prices of all the properties when they were sold last 
#(not all properties have a price because they have not been sold yet) 
df['PRICE'].describe()
```

```python
df['PRICE'].head()
```

```python
#This sorts the data by zipcode
df.sort_values('ZIPCODE').head()
```

```python
#This isolates and shows all the properties that have stories
hasStories = df['STORIES'].notnull()
df[hasStories].head()
```

```python
#This creates a new column indicating "Yes" if a property is mulitstory (>=2 stories) and "No" if it isn't
df["MULTISTORY"] = 'No'
df.loc[df['STORIES'] >= 2, "MULTISTORY"] = 'Yes'
df[hasStories].head(20)
```

```python
#Creats a new variable of just the properties that are not multistory (surprisingly the average is above 1)
noStories = df.loc[df['MULTISTORY'] == 'No']
noStories.describe()
```

```python
#The percent of the properties that are not multistory
percentNoStory = (len(noStories)/len(df))*100
percentNoStory
```

```python
#Creates a new column with the total number of bathrooms
df["Total Bathroom"] = df["BATHRM"] + df["HF_BATHRM"]
df.head(20)
```

```python
df['PRICE'].describe()
```

```python
#How many total bathrooms are there on average for the properties in the uppermost quartile (75%-max) of PRICE
expensiveProp = df.loc[(df['PRICE'] >= 6.520000e+05) & (df['PRICE'] <= 1.374275e+08)]
expensiveProp.describe()
```

```python
#How many total bathrooms are there on average for the properties in the lowermost quartile (min-25%) of PRICE
cheaperProp = df.loc[(df['PRICE'] <= 2.400000e+05)]
cheaperProp.describe()
```

```python
#What is the price of the most expensive property that does not have a single bathroom?
noBath = df.loc[df['Total Bathroom'] == 0]
noBath.describe()
```

```python
#Finds all the properties without bathrooms that have a price
hasPrice = noBath['PRICE'].notnull()
noBath[hasPrice].describe()
```

```python
pricesDF = noBath[hasPrice]
pricesDF['PRICE'].describe()
```

```python
#Creating bins for the prices of all the data based on the 1st quartile, 2nd quartile, 3rd quartile, 4th quartile, and max 
bins = [0, 2.400000e+05, 3.999990e+05, 6.520000e+05, 1.374275e+08]


group_names = ['Cheaper', 'A Little More Expensive', 'More Expensive', 'Most Expensive']


df['PRICE GRADE'] = pd.cut(df['PRICE'], bins, labels=group_names)
df
```
